```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
locale_df = pd.read_csv('locale_downloads_data/webstats-tb-locale-2020-12-27-2021-05-28.csv', skiprows=5)
```

```python
locale_df.locale.value_counts()
```

```python
ukrainian_df = locale_df[locale_df.locale.str.contains('uk')]
ukrainian_df.date.value_counts()
```

```python
russian_df = locale_df[locale_df.locale.str.contains('ru')]
russian_df.date.value_counts()
```

```python
fixed_russian_df = russian_df[~russian_df.locale.str.contains('ru-RU')]
fixed_russian_df
```

```python
english_us_df = locale_df[locale_df.locale == 'en-US']
english_us_df.date.value_counts()
```

```python
eng_pivot_df = english_us_df.pivot(index='date', 
                                   columns='locale', 
                                   values=['initial_downloads', 'update_pings'])
eng_pivot_df
```

```python
rus_pivot_df = fixed_russian_df.pivot(index='date', 
                                   columns='locale', 
                                   values=['initial_downloads', 'update_pings'])
rus_pivot_df
```

```python
en_and_rus_df = rus_pivot_df.merge(eng_pivot_df, on='date', how='inner')
en_and_rus_df
```

```python
en_and_rus_df.columns = en_and_rus_df.columns.map('_'.join)
```

```python
en_and_rus_df.to_csv('ru-en-US_downloads_updates_data_Dec-May.csv', index=True)
```
