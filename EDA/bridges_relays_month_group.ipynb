```python
import pandas as pd
import numpy as np
```

```python
rb_2022_df = pd.read_csv('datasets/ukr_rus_ir_all_bridge_relay_2011-2022.csv')
```

```python
rb_2021_df = pd.read_csv('datasets/ukr_rus_ir_all_bridge_relay_2020-2021.csv')
```

```python
rb_2021_2022_df = pd.concat([rb_2021_df, rb_2022_df]).reset_index().drop(columns=['index'])
rb_2021_2022_df
```

```python
rb_2021_2022_df.date = pd.to_datetime(rb_2021_2022_df.date, format='%Y-%m-%d')
```

```python
rb_2021_2022_df.info()
```

```python
rb_2021_2022_df['year'] = rb_2021_2022_df['date'].dt.year
rb_2021_2022_df['month'] = rb_2021_2022_df['date'].dt.month
```

```python
year_month_grouped = rb_2021_2022_df.groupby(['year', 'month'], as_index=False).agg('mean')
year_month_grouped
```

```python
year_month_grouped.to_csv('relay_bridge_2020-2022_group_year_month.csv', index=False)
```
